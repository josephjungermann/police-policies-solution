```python
mo_list = pd.read_csv("../output/MO_FatalEncounters.csv")
```

```python
mo_list.head()
```

```python
mo_list.info()
```

```python
# There are numerous columsn like "Date (Year)" listed as an integer... 
# ...we do not want to add these as numbers so they must be changed to string objects.
mo_list['Date (Year)'] = mo_list['Date (Year)'].astype(str);
mo_list['Location of death (zip code)'] = mo_list['Location of death (zip code)'].astype(str);
mo_list['Unique ID'] = mo_list['Unique ID'].astype(str);
mo_list['Latitude'] = mo_list['Latitude'].astype(str);
mo_list['Longitude'] = mo_list['Longitude'].astype(str);
mo_list['Video'] = mo_list['Video'].astype(str);
mo_list['Unique ID formula'] = mo_list['Unique ID formula'].astype(str);
mo_list["Subject's age"] = mo_list["Subject's age"].astype(str);
mo_list.info()
```

```python
# Find out what types of fatal encounters there are by looking into the "Cause of Death" column
mo_list["Cause of death"].value_counts()
```

```python
# Filter the list to only include and focus on fatal encounters that were "gunshot" related.
# We can focus on other causes of fatal encounters like "Tasered" and "Asphyxiated/Restrained" in a follow-up.
mo_cause = "Gunshot"
```

```python
# We should get 222 rows per our .value_counts() formula above.
mo_gunshot = mo_list[mo_list["Cause of death"] == mo_cause]
mo_gunshot
```

```python
# Fill any NaN columns with blank space for data-cleaning purposes.
mo_gunshot.fillna("")
```

```python
# Now that we have a dataset for gunshot killings only...
# ...next, we'll see how many shootings there were per year for MO as a whole, sort values and reset index.
mo_gunshot.info()
```

```python
# Are there any discrepencies with the data?.. 
# ...specifically mentioned by the Disposition and Intentional Use of Force Columns?
mo_gunshot["Intentional Use of Force (Developing)"].value_counts()
```
