```python
ks_list = pd.read_csv("../output/KS_FatalEncounters.csv")
```

```python
ks_list.head()
```

```python
ks_list.info()
```

```python
# There are numerous columsn like "Date (Year)" listed as an integer... 
# ...we do not want to add these as numbers so they must be changed to string objects.
ks_list['Date (Year)'] = ks_list['Date (Year)'].astype(str);
ks_list['Location of death (zip code)'] = ks_list['Location of death (zip code)'].astype(str);
ks_list['Unique ID'] = ks_list['Unique ID'].astype(str);
ks_list['Latitude'] = ks_list['Latitude'].astype(str);
ks_list['Longitude'] = ks_list['Longitude'].astype(str);
ks_list['Video'] = ks_list['Video'].astype(str);
ks_list['Unique ID formula'] = ks_list['Unique ID formula'].astype(str);
ks_list.info()
```

```python
# Find out what types of fatal encounters there are by looking into the "Cause of Death" column
ks_list["Cause of death"].value_counts()
```

```python
# Filter the list to only include and focus on fatal encounters that were "gunshot" related.
# We can focus on other causes of fatal encounters like "Tasered" and "Asphyxiated/Restrained" in a follow-up.
ks_cause = "Gunshot"
```

```python
# We should get 222 rows per our .value_counts() formula above.
ks_gunshot = ks_list[ks_list["Cause of death"] == ks_cause]
ks_gunshot
```

```python
# Fill any NaN columns with blank space for data-cleaning purposes.
ks_gunshot.fillna("")
```

```python
# Now that we have a dataset for gunshot killings only...
# ...next, we'll see how many shootings there were per year for KS as a whole, sort values and reset index.
ks_gunshot.info()
```

```python
# Are there any discrepencies with the data?.. 
# ...specifically mentioned by the Disposition and Intentional Use of Force Columns?
ks_gunshot["Intentional Use of Force (Developing)"].value_counts()
```

** Keep in mind that out of these 222 gunshot deaths, 75 were suicides, according to the data **
